# read_wholeday_v2.ipynb — Whole-day Type III Burst Count 

Same task as `read_wholeday_v1.ipynb` with four targeted improvements.

| # | Improvement | v1 behaviour | v2 behaviour |
|---|-------------|-------------|-------------|
| 1 | `get_sunrise_sunset` speed | `get_sun()` on every sample (slow) | Sparse sampling — 100 points, fast |
| 2 | `preproc2` scope | Called once on full file — global background | Called per-hour — local background |
| 3 | Burst count | Raw `len(line_sets)` — includes single-line groups | `sum(len(ls)>1)` — confirmed bursts only |
| 4 | Hour coverage | Daytime mask excludes night hours (always 0) | All 24 hours processed; daytime shaded on plot |
| 5 | Visualisation | No per-hour spectrum | `imshow` per hour for visual verification |


## 1  Imports

```python
import matplotlib.dates as mdates
import datetime
import matplotlib.pyplot as plt
import numpy as np
import os
import scipy.ndimage
import DetectRadioburst_OSRA as drb

from DetectRadioburst_OSRA import append_daily_csv

from astropy.time import Time
from astropy.coordinates import EarthLocation, get_sun, AltAz
import astropy.units as u
```

## 2  Load data

```python
# ── File path — first configurable line ──────────────────────────────────────
fname = "/net/lyot/scratch3/vocks/OSRA/2003/CD_300/031027_300.roh"

# Load full CD (day + night)
dyspec, t_fits, f_fits = drb.read_osraf2(fname)
t_fits = drb.to_python_datetime(t_fits)   #time conversion

print(f"Loaded {len(t_fits)} records")
print(f"Time range : {t_fits[0]}  ->  {t_fits[-1]}")
```

## 3  Sunrise / Sunset — Tremsdorf
> **Improvement 1:** sparse sampling — `get_sun()` called on ~100 points instead of every record.

```python
# ── Tremsdorf observatory coordinates ─────────────────────────────────────────
location = EarthLocation(
    lat=52.2167 * u.deg,
    lon=13.1833 * u.deg,
    height=65 * u.m
)

# ── Sunrise / sunset — IMPROVEMENT 1: sparse sampling (fast) ─────────────────
# v1 called get_sun() on every time sample (slow).
# v2 uses every Nth point — 100 samples is enough to locate the zero-crossing.
def get_sunrise_sunset(t_fits, location, step=None):
    if step is None:
        step = max(1, len(t_fits) // 100)
    t_sparse  = t_fits[::step]
    times     = Time(list(t_sparse))
    altaz_frame = AltAz(obstime=times, location=location)
    sun_altaz = get_sun(times).transform_to(altaz_frame)
    altitudes = sun_altaz.alt.deg + 0.833   # refraction correction

    sunrise_idx = np.where((altitudes[:-1] < 0) & (altitudes[1:] >= 0))[0]
    sunset_idx  = np.where((altitudes[:-1] >= 0) & (altitudes[1:] < 0))[0]

    sunrise_time = t_sparse[sunrise_idx[0]] if len(sunrise_idx) > 0 else t_fits[0]
    sunset_time  = t_sparse[sunset_idx[0]]  if len(sunset_idx)  > 0 else t_fits[-1]
    return sunrise_time, sunset_time

sunrise, sunset = get_sunrise_sunset(t_fits, location)

sunrise = drb.to_python_datetime([sunrise])[0] # conversion  sunrise
sunset  = drb.to_python_datetime([sunset])[0]    # conversion of sunset

print("Sunrise :", sunrise)
print("Sunset  :", sunset)
```

## 4  Daytime mask — annotation only
> **Improvement 4:** mask is computed but **not** used to filter the loop. All 24 hours are processed. The mask shades the final plot.

```python
# ── Daytime mask — kept for annotation, NOT applied as data filter ────────────
# IMPROVEMENT 4: v1 excluded night hours entirely so they always showed 0 bursts.
# v2 processes all 24 hours; sunrise/sunset are used only to shade the final plot.
time_mask_day = (t_fits >= sunrise) & (t_fits <= sunset)
day_t_fits = t_fits[time_mask_day]
print(f"Daytime samples : {np.sum(time_mask_day)} / {len(t_fits)}")
print("Note: mask used for plot annotation only — all 24 hours are processed.")
```

## 5  Storage arrays

```python
# ── Storage arrays — one entry per hour ──────────────────────────────────────
bursts_per_hour   = np.zeros(24, dtype=int)   # confirmed bursts (groups with >1 line)
len_mask_per_hour = np.zeros(24, dtype=int)   # time samples in that hour
```

## 6  Hour loop — 00:00 to 23:00 UT
> **Improvement 2:** `preproc2` called per-hour.  
> **Improvement 3:** burst count uses `len(ls) > 1` gate.  
> **Improvement 5:** per-hour `imshow` for visual verification.

```python
#  FIX TIME TYPE BEFORE LOOP
t_fits = np.array(t_fits).astype('datetime64[ms]').astype(datetime.datetime)
base_date = t_fits[0].replace(hour=0, minute=0, second=0, microsecond=0)

# ── Hour loop 0..23 ───────────────────────────────────────────────────────────
results =[]
for h in range(24):
    start_time = datetime.datetime(
        t_fits[0].year, t_fits[0].month, t_fits[0].day, h, 0, 0
    )
    end_time = start_time + datetime.timedelta(hours=1)

    # Hour mask only — no daytime restriction
    time_mask = (day_t_fits >= start_time) & (day_t_fits < end_time)

    idx = np.where(time_mask)[0]
    len_mask_per_hour[h] = len(idx)

    if len(idx) == 0:
        bursts_per_hour[h] = 0
        print(f"Hour {h:02d}: no data")
        continue

    data_cut   = dyspec[idx, :]
    t_cut      = t_fits[idx]

    
    
    # preproc2 per-hour cutout ──────────────────────────────────
   
    # v2 normalises each hour independently — same approach as read_osraf2nightpro.
    data_fits_new, data_fits_new_smooth = drb.preproc2(data_cut, gauss_sigma=(5.5, 0.0))

    # vmin/vmax for imshow —                                       
    data_safe_arr = data_fits_new_smooth.ravel()
    data_safe = np.sort(data_safe_arr)[
        int(data_safe_arr.shape[0] * 0.02):int(data_safe_arr.shape[0] * 0.98)]
    vmin = np.nanmean(data_safe) - 2  * np.nanstd(data_safe)
    vmax = np.nanmean(data_safe) + 4  * np.nanstd(data_safe) + 0.5 * np.nanmax(data_safe)

    # per-hour spectrum imshow ──────────────────────────────────
    t_plot = mdates.date2num(t_cut)
    fig, ax = plt.subplots(figsize=(8, 3))
    ax.imshow(data_fits_new_smooth.T, aspect='auto', origin='lower',
              vmin=vmin, vmax=vmax, cmap='inferno',
              extent=[t_plot[0], t_plot[-1], f_fits[0], f_fits[-1]])
    ax.xaxis_date()
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
    ax.set_xlabel('Time (UT)')
    ax.set_ylabel('Frequency (MHz)')
    ax.set_title(f'Dynamic spectrum  {h:02d}:00 – {h+1:02d}:00 UT')
    plt.tight_layout()
    plt.show()

    # Binarization
    bmap = drb.binarization(data_fits_new_smooth, N_order=8, peak_r=0.9993)

    # Hough detection
    lines = drb.hough_detect(
        bmap, data_cut,
        threshold=30, line_gap=10, line_length=50,
        theta=np.linspace(np.pi/2 - np.pi/8, np.pi/2 - 0.5/180*np.pi, 300)
    )

    # Group lines (empty-safe — already edited into drb)
    line_sets = drb.line_grouping(lines, min_dist=10)

    #  count only groups with >1 line ────────────────────────────
    # A Type III burst needs at least two Hough lines to define a drift track in most cases.
    confirmed = sum(1 for ls in line_sets if len(ls) > 1)

    bursts_per_hour[h] = confirmed

    results.append({
        "hour": h,
        "bursts": confirmed,
        "raw_groups": len(line_sets),
        "samples": len(idx),
    })
    print(f"Hour {h:02d}: bursts={confirmed:>3} (raw groups={len(line_sets):>3}),  "
          f"samples={len(idx):>6}")
```

```python
# ── Results ───────────────────────────────────────────────────────────────────
print("\nBursts per hour    :", bursts_per_hour)
print("Samples per hour   :", len_mask_per_hour)

# ── Determine daytime hours from the sunrise/sunset ────────────────────────────
sunrise_h = sunrise.hour + sunrise.minute / 60
sunset_h  = sunset.hour  + sunset.minute  / 60

# ── Plot ──────────────────────────────────────────────────────────────────────
fig, ax1 = plt.subplots(figsize=(13, 4))
ax2 = ax1.twinx()

ax1.bar(range(24), bursts_per_hour, color='steelblue', alpha=0.75,
        label='Confirmed bursts/hour  (groups > 1 line)')
ax2.plot(range(24), len_mask_per_hour, 'r--o', ms=4, lw=1.2, label='Samples/hour')

# Shade daytime
ax1.axvspan(sunrise_h, sunset_h, color='gold', alpha=0.15, label=f'Daytime ({sunrise.strftime("%H:%M")}–{sunset.strftime("%H:%M")} UT)')

ax1.set_xlabel('Hour (UT)')
ax1.set_ylabel('Number of confirmed burst groups', color='steelblue')
ax2.set_ylabel('Number of time samples', color='red')
ax1.set_xticks(range(24))
ax1.set_xlim(-0.5, 23.5)
ax1.set_title('Whole-day Type III burst count  —  031027 CD_300  '
              '(v2: all 24 h, per-hour preproc, confirmed bursts only)')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left', fontsize=8)
plt.tight_layout()
plt.show()
```
